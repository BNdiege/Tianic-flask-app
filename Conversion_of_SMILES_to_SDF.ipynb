```python
# --- Install RDKit in Colab ---
!pip install rdkit

import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
# --- Step 1: Load your CSV ---
# Suppose your file is called "compounds.csv" and has a column "SMILES"
df = pd.read_csv("PE_trial conversion.csv", encoding='latin1')
df
```

```python
# If your SMILES column is named differently, change "SMILES" to your column name
smiles_list = df["SMILES"].tolist()
```

```python
# --- Step 2: Convert SMILES to RDKit molecules ---
mols = []
for smi in smiles_list:
    mol = Chem.MolFromSmiles(smi)
    if mol is not None:  # skip invalid SMILES
        mol = Chem.AddHs(mol)  # add hydrogens
        AllChem.EmbedMolecule(mol, AllChem.ETKDG())  # generate 3D coordinates
        mols.append(mol)
```

```python
# --- Step 3: Write molecules to SDF file ---
writer = Chem.SDWriter("compounds.sdf")
for mol in mols:
    writer.write(mol)
writer.close()

print(f"Successfully converted {len(mols)} compounds to SDF!")
```
